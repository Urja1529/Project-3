```python
import pandas as pd
import re

import requests
import json
from pprint import pprint
import time
```

```python
with open('met_data.json') as json_file:
    data = json.load(json_file)
    print(type(data))
    print(data[10])
```

```python
all_data = []

with open('met_data.json') as json_file:
    data = json.load(json_file)
    index = 0
    for art in data:
        try:
            temp_url = f'https://collectionapi.metmuseum.org/public/collection/v1/objects/{art["objectID"]}'
            temp_response = requests.get(temp_url).json()
            all_data.append(temp_response)

            time.sleep(0.01) # less than 70 calls/second
            if index % 1000 == 0:
                print(index)
            index += 1
        except:
            break

print("----------complete-----")
```

```python
data_df = pd.DataFrame(all_data)
data_df.head(5)
```

```python
# Clean cat data
data_clean_df = data_df[['objectBeginDate', 'objectEndDate', 'title', 'objectName', 'medium', 'primaryImage', 'artistDisplayName', 'artistDisplayBio']]
data_clean_df = data_clean_df.replace(r'^s*$', float('NaN'), regex = True)
data_clean_df.dropna(inplace = True) 
data_clean_df.head(5)
```

```python
# Transfrom the data for timeline

data_timeline = pd.DataFrame(columns = ['Year', 'Month', 'Day', 'Time', 'End Year', 'End Month', 'End Day', 'End Time', 'Display Date', 'Headline', 'Text', 'Media', 'Media Credit', 'Media Caption', 'Media Thumbnail', 'Type', 'Group', 'Background'])

data_timeline['Year'] = data_clean_df['objectBeginDate']
data_timeline['End Year'] = data_clean_df['objectEndDate']
data_timeline['Headline'] = data_clean_df[['title', 'objectName', 'medium']].agg(' '.join, axis=1)
data_timeline['Text'] = data_clean_df['medium']
data_timeline['Media'] = data_clean_df['primaryImage']
data_timeline['Media Credit'] = data_clean_df['artistDisplayName']
data_timeline['Media Caption'] = data_clean_df['artistDisplayBio']


data_timeline.head(5)

data_timeline.to_csv('data_timeline.csv')
```
